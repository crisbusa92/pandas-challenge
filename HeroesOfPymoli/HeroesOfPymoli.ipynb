## Purchasing Analysis (Age)

```python
auq = file.drop_duplicates("SN")
bins = [0,9.99,14,19,24,29,34,39,46]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
auq["Age Table"] = pd.cut(auq["Age"], bins, labels=group_names, include_lowest=True)
ac = auq["Age Table"].value_counts()
ap = ac/len(auq)
p = pd.DataFrame({"Count by Age":ac, "Percentage of Players":ap})
p.sort_index()



```
